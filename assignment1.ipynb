Q3

```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
valX,countX = np.unique(x,return_counts=True)
mostFreqX = valX[np.argmax(countX)]
indicesX = np.where(x == mostFreqX)[0]
print(indicesX)
print(mostFreqX, end = '\n\n')

y = np.array([1,1,1,2,3,4,2,4,3,3,])
valY,countY = np.unique(y,return_counts=True)
mostFreqY = valY[np.argmax(countY)]
indicesY = np.where(y == mostFreqY)[0]
print(indicesY)
print(mostFreqY, end = '\n  ')
```
